We give you the network of the 34 members of [karate clubs](https://en.wikipedia.org/wiki/Zachary's_karate_club) on the campus. Each node is a member of one of two clubs ("Mr. Hi" or "Officer") and there is an edge between two nodes if the two are friends outside of the club.

```python
karate_club = nx.karate_club_graph()
```

```python
# Your code goes here

# nodos_tipo = [node['club'] for node in karate_club.nodes]
# color_map = (lambda node: 'blue' if karate_club.nodes[node]['club']=='Mr. Hi' else 'red', karate_club.nodes)

plt.figure('figure1',figsize=(10,10))
color_map = []
for node in karate_club.nodes:
    color_map.append((lambda node: 'lightblue' if karate_club.nodes[node]['club']=='Mr. Hi' else 'red')(node))

nx.draw_networkx(karate_club, node_color=color_map, with_labels=True)
plt.axis('off')
```
